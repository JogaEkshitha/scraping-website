```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/naukri_jobs.csv")

# Display first few rows
df.head()
```

```python
print(df.info())  # Check data types and null values
print(df.head())  # View the first few rows
```

```python
print(df.isnull().sum())
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.head(5)
```

```python
from IPython.core.display import display, HTML

# Function to display aligned DataFrame with fixed headers
def show_aligned_df(df):
    styled_df = df.style.set_properties(**{'text-align': 'left'})\
                        .set_table_styles([{'selector': 'th', 'props': [('text-align', 'left')]}])
    display(HTML(styled_df.to_html()))

# Display first 5 rows with proper alignment
show_aligned_df(df.head(5))



```

```python
show_aligned_df(df.head(50))
```

```python
df[['min_experience', 'max_experience']] = df['experience'].str.extract(r'(\d+)-(\d+)').astype(float)
df['experience'] = df[['min_experience', 'max_experience']].mean(axis=1)
df.drop(columns=['min_experience', 'max_experience'], inplace=True)  # Remove temp columns
```

```python
show_aligned_df(df.head(5))
```

```python
df['experience'] = df['experience'].map('{:.0f}'.format)
show_aligned_df(df.head(50))
```

```python
df[['min_salary', 'max_salary']] = df['salary'].str.extract(r'(\d+)-(\d+)').astype(float)
df['salary'] = df[['min_salary', 'max_salary']].mean(axis=1) * 1_00_000  # Convert Lacs to Absolute
df.drop(columns=['min_salary', 'max_salary'], inplace=True)  # Remove temp columns
```

```python
show_aligned_df(df.head(5))
```

```python
df['salary'] = df['salary'].fillna(0).astype(int)
show_aligned_df(df.head(50))
```

```python
print(df.isnull().sum())
```

```python
df['location'] = df['location'].apply(lambda x: x.split(',')[0])
```

```python
show_aligned_df(df.head(5))
```

```python
df['job_title'] = df['job_title'].apply(lambda x: x.split(',')[0])
```

```python
show_aligned_df(df.head(10))
```

```python
print(df.duplicated().sum())  # Count of duplicate rows
print(df[df.duplicated()])    # Display duplicate rows
```

```python
print(df.duplicated().sum())
```

```python
print(df[df.duplicated()])
```

```python
df.drop_duplicates(keep='first', inplace=True)
```

```python
print(df.duplicated().sum())
```

```python
df.to_csv('cleaned_data.csv', index=False)
print(" Data Cleaning Completed! Ready for Dashboard ")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your cleaned DataFrame (assuming it's already cleaned)
# df = pd.read_csv('your_cleaned_data.csv')  # Uncomment if reading from a CSV

### 1️⃣ Salary Distribution - Histogram & Boxplot ###
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(df['Salary'], bins=20, kde=True, color='blue')
plt.title('Salary Distribution (Histogram)')
plt.xlabel('Salary (LPA)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.boxplot(x=df['Salary'], color='green')
plt.title('Salary Distribution (Boxplot)')

plt.show()

### 2️⃣ Top Companies Hiring the Most Data Scientists ###
top_companies = df['Company'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(y=top_companies.index, x=top_companies.values, palette='viridis')
plt.xlabel('Number of Job Listings')
plt.ylabel('Company')
plt.title('Top 10 Companies Hiring Data Scientists')
plt.show()

### 3️⃣ Salary vs. Experience - Scatter Plot & Correlation ###
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Experience'], y=df['Salary'], color='red')
plt.title('Salary vs. Experience')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary (LPA)')
plt.show()

# Correlation
corr = df[['Experience', 'Salary']].corr()
print("Correlation between Experience and Salary:\n", corr)

### 4️⃣ Job Locations with the Highest Opportunities ###
top_locations = df['Location'].value_counts().head(10)

plt.figure(figsize=(10, 5))
sns.barplot(y=top_locations.index, x=top_locations.values, palette='coolwarm')
plt.xlabel('Number of Job Listings')
plt.ylabel('Location')
plt.title('Top 10 Job Locations for Data Scientists')
plt.show()
```
